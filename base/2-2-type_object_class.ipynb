## type、object 和 class 的关系

### type 生成了 class， class 生成了 obj
- type 是用来生成对象的
- 1 是使用内建的类的实例

```python
# type -> int -> 1
a = 1
print("type(1) = ", type(1))
print("type(a) = ", type(a))
print("type(int) = ", type(int))

# type -> class -> obj
print("")
b = "abc"
print("type(b) = ", type(b))
print("type(str) = ", type(str))
```

### type 生成了自定义类，自定义类生成了自定义对象   

```python
class Student:
    pass

class MyStudent(Student):
    pass

str = Student()
print("type(str) = ", type(str))
print("type(Student) = ", type(Student))
```

### 常用的内置对象

```python
a = [1, 2]
print("type(a) = ", type(a))
print("type(list) = ", type(list))
```

### Object 是所有对象的基类
- Object 是最顶层基类
- type 也是一个类

```python
class Student:
    pass
class MyStudent(Student):
    pass

print("Student.__bash__ = ", Student.__base__)
print("MyStudent.__bash__ = ", MyStudent.__base__)

print("type.__base__ = ", type.__base__)
print("type(object) = ", type(object))
print("object.__bases__ = ", object.__bases__)
```
