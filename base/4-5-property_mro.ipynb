## 类属性和实例属性的查找顺序 MRO

```python
class A:
    name = "A"
    def __init__(self):
        self.name = "obj"

a = A()
print("a.name = ", a.name)
```

### MRO 算法
- python2 使用：深度优先检索
    - 对于菱形继承，则会出现问题
- python2.3 使用：广度优先检索
    - 如果 D 和 C 存在同名函数，则 C 会覆盖 D 的方法，会出现问题
- python2.3 之后：使用 C3 算法

```python
class D(object):
    pass
class C(D):
    pass
class B(D):
    pass
class A(B,C):
    pass
print("A.__mro__ = ", A.__mro__)
```
